```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.options.display.max_columns = None
zomato_df = pd.read_csv(r'E:\data analysis projects\Zomato Data Set Analysis & Visualization\cleaned_zomato.csv')
```

```python
zomato_df
```

# online_order or not 

```python
plt.figure(figsize=(5,5))

sns.countplot(
    data=zomato_df,
    x='online_order',
    width=.2,
    palette = 'inferno'
)
plt.title("Restaurants delivering Online or not")
plt.legend()
```

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>
        number of resturants that has online more than  resturants that hasn't online
    </span>    
</div>

```python
plt.figure(figsize=(5,5))
sns.displot(
    data=zomato_df,
    x='online_order',
    y='rate',
    rug=True,
    palette = 'inferno'
)
plt.title("Restaurants delivering Online or not VS rating")
plt.legend()
```

# Restaurants allowing table booking or not

```python
table_count = zomato_df["book_table"].value_counts()

plt.pie(table_count.values, labels=table_count.index, autopct="%1.0f%%")
plt.title("Restaurants allowing table booking or not")
plt.show
```

```python
plt.figure(figsize=(5,5))

sns.barplot(
    data=zomato_df,
    x='book_table',
    y='rate',
    width=.2,
)
plt.title("Restaurants allowing table booking or not VS rate")
plt.legend()
```

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>
        resturants that has table booking has the higher rate 
    </span>    
</div>

# Best locations

```python
Locations_cities = pd.DataFrame(zomato_df.groupby('location')['rate'].mean())
Locations_cities['location_count'] = zomato_df['location'].value_counts()
top20 = Locations_cities.sort_values(by='rate' ,ascending=False).head(20)
top20
```
