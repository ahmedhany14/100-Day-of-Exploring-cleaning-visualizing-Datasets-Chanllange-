```python
import pandas as pd
import matplotlib.pyplot  as plt
import numpy as np
import seaborn as sns
import collections

df = pd.read_csv(r"E:\datasets\dataset.csv")
df
```

```python
df.columns = df.columns.str.lower()
```

```python
columns = list(df.columns)
columns
```

# renaming columns name

```python
new_names = collections.defaultdict(list)
for i in columns:
    before,after = i, i
    if i.count(' '):
        after = after.replace(' ','_')
    new_names[before] = after
df.rename(columns=new_names,inplace=True)
df.columns
```

# spliting full name to first and last

```python
df['first_name'] = df['full_name'].str.split(' ').str.get(0)
df['last_name'] = df['full_name'].str.split(' ').str.get(1)
df
```

# Dropping full name and insert first and last at the front

```python
first_name = df["first_name"]
last_name = df["last_name"]
df.drop(columns=["full_name", "first_name", "last_name"], inplace=True)
df.insert(1, column="first_name", value=first_name)
df.insert(2, column="last_name", value=last_name)
df
```

# Convert names to lower

```python
df['first_name'] = df['first_name'].str.lower()
df['last_name'] = df['last_name'].str.lower()
df['first_name'] = df['first_name'].str.strip()
df['last_name'] = df['last_name'].str.strip()
```

# convert income to int 

```python
df['income.1'].dtype
```

```python

def convet_to_integer():
    items = df['income.1']
    integer_items = []
    for i in items:
        numeric = ''
        for j in i:
            if j == '.': break
            if j == '$' :continue
            k = str(j)
            numeric += k
        if numeric == '':
            integer_items.append(0)
        else :
            integer_items.append(int(numeric))
    return integer_items

df['New_income_in_doller'] = convet_to_integer()
df
```

```python
df.drop(columns='income.1', inplace=True)
df
```
