```python
df = netflix_bestMovies[['DURATION', 'SCORE', 'MAIN_GENRE']]
print(df['DURATION'].min(), df['DURATION'].max())
def Categorize(Time):
    if Time > 20 and Time < 40:
        return '20:40'
    if Time >= 40 and Time < 100:
        return '40:100'
    if Time >= 100 and Time < 150:
        return '100:150'
    if Time > 150 and Time < 200:
        return '150:200'
    else:
        return 'over 200'
    
df['DURATION_Categorize'] = df['DURATION'].apply(Categorize)
df
```

```python
pivot = df.pivot_table(index='DURATION_Categorize',columns='MAIN_GENRE',values='SCORE', fill_value=0)
pivot.sort_values(by='DURATION_Categorize' ,ascending=False, inplace=True)
pivot
```

```python
sns.heatmap(
    pivot,
    annot=True
)
```

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>
        all films that have DURATION in intrval 100 : 150 has strog correlation with Score
