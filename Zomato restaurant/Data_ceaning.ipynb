```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
pd.options.display.max_columns = None
zomato_df = pd.read_csv(r'E:\data analysis projects\Zomato Data Set Analysis & Visualization\zomato.csv')
new_zomato_df = zomato_df
```

```python
zomato_df.head(5)
```

<a id="intro"></a>
<h2>   
      <font color = blue >
            <span style='font-family:Georgia'>
            1. delete nuneeded columns (url, address, phone, reviews_list, menu_item, dish_liked):
            </span>   
        </font>    
</h2>

```python
new_zomato_df.drop(columns=['url', 'address', 'phone', 'reviews_list', 'menu_item', 'dish_liked'], inplace=True)
new_zomato_df.head(5)
```

<a id="intro"></a>
<h2>   
      <font color = blue >
            <span style='font-family:Georgia'>
            2. rename columns:
            </span>   
        </font>    
</h2>

```python
new_zomato_df.rename(columns={'rest_type': 'resturant_type', 'listed_in(city)':'city' ,'approx_cost(for two people)':'cost_for_2','listed_in(type)':'type'}, inplace=True)
new_zomato_df
```

<a id="intro"></a>
<h2>   
      <font color = blue >
            <span style='font-family:Georgia'>
            3. Data Type conversion, and fix cost and rate column:
            </span>   
        </font>    
</h2>

<h3 name='libraries'>   
      <font color = purple >
            <span style='font-family:Georgia'>
            3.1 fix and convert rate to float instead of object:
            </span>   
        </font>    
</h3>

```python
# fixing rating

def fix_rate(rate):
    if rate == 'NEW' or rate == '-' or rate == np.nan:
        return np.nan
    else:
        new_rate = str(rate).split('/')
    return float(new_rate[0])

new_zomato_df['rate'] = new_zomato_df['rate'].apply(fix_rate)

new_zomato_df['rate']
```

<h3 name='libraries'>   
      <font color = purple >
            <span style='font-family:Georgia'>
            3.2 fix and convert cost to float instead of object:
            </span>   
        </font>    
</h3>

```python
# fixing cost

def fix_cost(cost):
    cost = str(cost).replace(',','')
    return float(cost)

new_zomato_df['cost_for_2'] = zomato_df['cost_for_2'].apply(fix_cost)

new_zomato_df['cost_for_2']
```

```python
new_zomato_df
```

<a id="intro"></a>
<h2>   
      <font color = blue >
            <span style='font-family:Georgia'>
            4. categorize any location that has less than 1000 as a other:
            </span>   
        </font>    
</h2>

```python
count_locations = new_zomato_df['location'].value_counts() 
locations_below1000 = count_locations[count_locations < 1000]
def fix_location(location):
    if location in locations_below1000:
        return 'others'
    return location

new_zomato_df['location'] = new_zomato_df['location'].apply(fix_location)
new_zomato_df['location'].unique()
```

<a id="intro"></a>
<h2>   
      <font color = blue >
            <span style='font-family:Georgia'>
            5. Null Value Data Imputation:
            </span>   
        </font>    
</h2>

```python
null_columns = []
rows = len(new_zomato_df)
for col in new_zomato_df:
    if new_zomato_df[col].isna().sum() > 0:
        null_columns.append([col, new_zomato_df[col].isna().sum() / rows * 100])
null_columns
```

<div class="alert alert-block alert-info">
    <span style='font-family:Georgia'>
        <b>Insight: </b><br>
        there is 3 object columns has missing values less than 1% so, you will fill them by mode, and the other column (rate), will fill by using class method, and the column cost_for_2 will fill by mean
    </span>    
</div>

<h3 name='libraries'>   
      <font color = purple >
            <span style='font-family:Georgia'>
            5.1 fill columns with mode:
            </span>   
        </font>    
</h3>

```python
new_zomato_df['location'].fillna(new_zomato_df['location'].mode()[0], inplace=True)
new_zomato_df['resturant_type'].fillna(new_zomato_df['resturant_type'].mode()[0], inplace=True)
new_zomato_df['cuisines'].fillna(new_zomato_df['cuisines'].mode()[0], inplace=True)
```

```python
new_zomato_df.isna().sum()
```

```python
new_zomato_df['cost_for_2'].fillna(new_zomato_df['cost_for_2'].mean, inplace=True)
```

```python
new_zomato_df.isna().sum()
```

<h3 name='libraries'>   
      <font color = purple >
            <span style='font-family:Georgia'>
            5.2 filling rate column with class method, using categorize column type:
            </span>   
        </font>    
</h3>

```python
new_zomato_df['type'].unique()
```

```python
column = new_zomato_df['type'].unique()

for col in column:
    print(len(new_zomato_df[new_zomato_df['type'] == col]['rate']) ,(new_zomato_df[new_zomato_df['type'] == col]['rate'].mean()), ' ')
    
for col in column:
    new_zomato_df.loc[new_zomato_df['type'] == col, 'rate'] = new_zomato_df[new_zomato_df['type'] == col]['rate'].fillna(new_zomato_df[new_zomato_df['type'] == col]['rate'].mean())
    
```
